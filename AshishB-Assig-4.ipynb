```python
import csv
def fmean(IRIS, sepal_width):
    total = 0
    count = 0
    with open(IRIS, 'r') as csvfile:
        reader = csv.DictReader(csvfile)
        for row in reader:
            total += float(row[sepal_width])
            count+=1
        if count == 0:
            return 0
        else:
            return total/count
```

```python
my_mean = fmean('IRIS.csv', 'sepal_width')
print(my_mean)
```

```python
import csv
def fmedian(IRIS, sepal_width):
    values = []
    with open(IRIS, 'r') as csvfile:
        reader = csv.DictReader(csvfile)
        for row in reader:
            values.append(float(row[sepal_width]))
    n = len(values)
    num = sorted(values)
    if n%2 == 0:
        median1 = num[n//2]
        median2 = num[n//2-1]
        median = (median1 + median2) / 2
    else:
        median = num[n//2]
    return median
```

```python
my_median = fmedian('IRIS.csv', 'sepal_width')
print(my_median)
```

```python
import csv
import statistics
def fmode(IRIS, sepal_width):
    values = []
    with open(IRIS, 'r') as csvfile:
        reader = csv.DictReader(csvfile)
        for row in reader:
            values.append(float(row[sepal_width]))
        try:
            mode = statistics.mode((values))
        except statistics.StatisticsError:
            mode = None
        return mode
```

```python
my_mode = fmode('IRIS.csv', 'sepal_width')
print(my_mode)
```

```python
import csv
import statistics
def fvariance(IRIS, sepal_width):
    values = []
    with open(IRIS, 'r') as csvfile:
        reader = csv.DictReader(csvfile)
        for row in reader:
            values.append(float(row[sepal_width]))
        variance = statistics.variance(values)
        return variance
```

```python
my_variance = fvariance('IRIS.csv', 'sepal_width')
print(my_variance)
```
